```python
def get_data_with_pagination(api_url, headers):
    all_results = []
    next_cursor = None
    i = 0
    while True:
        print(f"Call {i}", end = "\t")
        i += 1
        # Make the API request with the current "next_cursor"
        params = {"start_cursor": next_cursor} if next_cursor else None
        print('\t', params)
        response = requests.post(api_url, headers=headers, json=params)
        if response.status_code == 200:
            data = response.json()
            current_results = data.get('results', [])
            print("Successful request. Got ", len(current_results), " results. \t\tCursor ", next_cursor)
        else:
            print("Failed to fetch data from Notion database")
            break

        # Process the current page of results
        all_results.extend(current_results)

        # Check if there are more results available
        print('\t', data.get('has_more'), data.get('next_cursor'))
        if data.get('has_more', False):
            next_cursor = data.get('next_cursor')
        else:
            next_cursor = None
            break

        #time.sleep(2)

    return all_results
```
