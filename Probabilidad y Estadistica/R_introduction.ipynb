```r
  graf_bastones = function(x, p){
    plot(x, p, pch=16, col = "slateblue") #pch es tipo de caracter
    for(linea in 1:length(x)){
      lines(c(x[linea], x[linea]), c(0, p[linea]), col="slateblue")
    }
  }
  x = 1:6
  p = rep(1/6, times = 6)
  graf_bastones(x, p)
```
