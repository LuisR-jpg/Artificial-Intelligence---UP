# Simple Linear Regression (Hours of study)

The goal of this project was to build a linear regression model from the ground up using scikitlearn.

```python
%matplotlib inline

#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

### Import the data

Here, we're using a dataset with two columns containing the amount of hours studied and the test scores students achieved, respectively.

```python
#Mount Drive
#from google.colab import drive
#drive.mount('/content/drive')

#Data path 
dataset_path = '/content/drive/MyDrive/Colab Notebooks/data.csv'
```

```python
#Dataset:
dataset = pd.read_csv(dataset_path)

#Extract columns
x = dataset.iloc[:, :-1].values
x = x.tolist()
y = dataset.iloc[:, 1].values
y = y.tolist()

#Plot the dataset
plt.scatter(x,y)
plt.xlabel('Hours of study')
plt.ylabel('Test scores')
plt.title('Dataset')
plt.show()
```

### Training the Algorithm

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn import datasets, linear_model
regressor = linear_model.LinearRegression()
regressor.fit(X_train,Y_train)
#print(regressor.intercept_)
#print(regressor.coef_)
```

```python
Y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})
df
```

```python
plt.scatter(x, y)
plt.plot(X_test, Y_pred, color='red', linewidth=3)
plt.xlabel('Hours of study')
plt.ylabel('Test scores')
plt.title('Line of best fit')
plt.show()
```

# Simple Linear Regression (California housing)

The goal of this project was to build a linear regression model from the ground up using scikitlearn.

```python
#Imports
import numpy as np
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt
```

### Import the data


```python
#Mount Drive
#from google.colab import drive
#drive.mount('/content/drive')

#Dataset
dataset = datasets.load_boston()
print("Dataset size:", dataset.data.shape)
```

```python
#Extract columns

#Select column "RM" (Average number of rooms per dwelling)
#print('Columns:',dataset.feature_names)
X = dataset.data[:, np.newaxis, 5]
Y = dataset.target
```

```python
#Plot the dataset
plt.scatter(X,Y)
plt.xlabel('Number of rooms')
plt.ylabel('Mean value')
plt.show()
```

### Training the Algorithm

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
regressor = linear_model.LinearRegression()
regressor.fit(X_train,Y_train)
#print(regressor.intercept_)
#print(regressor.coef_)
```

```python
Y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})
df
```

```python
plt.scatter(X, Y)
plt.plot(X_test, Y_pred, color='red', linewidth=3)
plt.title('Line of best fit')
plt.xlabel('Number of rooms')
plt.ylabel('Mean value')
plt.show()
```
