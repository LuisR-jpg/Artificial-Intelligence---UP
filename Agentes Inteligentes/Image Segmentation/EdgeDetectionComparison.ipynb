### Canny Edge Detection (*OpenCV*)

```python
edges = cv2.Canny(img.copy(),100,200)

plt.figure(figsize=(6,6))
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

### Ours

#### *Clean image*

```python
gray = img.copy()

a = 10

# blur
blur = cv2.GaussianBlur(gray, (0,0), sigmaX=a, sigmaY=a)
divide = cv2.divide(gray, blur, scale=50)
imgNormalize1 = cv2.normalize(divide,None,0,255,cv2.NORM_MINMAX)
thresh = cv2.threshold(divide, 0, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)[1]

# apply morphology
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
morph = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
# show result
plt.figure(figsize=(8,8))
plt.imshow(morph, cmap = "gray")
```

#### *High-pass filter*

```python
img_float32 = np.float32(morph)

dft = cv2.dft(img_float32, flags = cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)

rows, cols = morph.shape
crow, ccol = int(rows/2) , int(cols/2)     # center
r = 120
mask = np.ones((rows, cols, 2), np.uint8)
mask[crow-r:crow+r, ccol-r:ccol+r] = 0
fshift = dft_shift*mask
f_ishift = np.fft.ifftshift(fshift)
img_One = cv2.idft(f_ishift)
img_One = cv2.magnitude(img_One[:,:,0],img_One[:,:,1])
```

#### *Show results*

```python

plt.figure(figsize=(15,15))
plt.subplot(1,3,1)
plt.imshow(img_One,cmap = 'gray');
plt.title('Ours');
plt.subplot(1,3,2);
plt.imshow(img,cmap = 'gray');
plt.title('Original');
plt.subplot(1,3,3);
plt.imshow(edges,cmap = 'gray');
plt.title('Canny openCV');
```

a) Use Canny edge detection and your gradient images to identify the borders with more cornerness. Use kernels that highlight the tipical structure of a corner and get the pixels that match the best.

```python
kernel = np.array([
    [1, 1, 1],
    [0, 0, 1],
    [0, 0, 1]
])
oOne = cv2.morphologyEx(edges, cv2.MORPH_HITMISS, kernel)
np.unique(oOne, return_counts=True)
```

```python
kernel = np.array([
    [0, 0, 1],
    [0, 0, 1],
    [1, 1, 1]
])
oTwo = cv2.morphologyEx(edges, cv2.MORPH_HITMISS, kernel)
np.unique(oTwo, return_counts=True)
```

```python
kernel = np.array([
    [1, 0, 0],
    [1, 0, 0],
    [1, 1, 1]
])
oThree = cv2.morphologyEx(edges, cv2.MORPH_HITMISS, kernel)
np.unique(oThree, return_counts=True)
```

```python
kernel = np.array([
    [1, 1, 1],
    [1, 0, 0],
    [1, 0, 0]
])
oFour = cv2.morphologyEx(edges, cv2.MORPH_HITMISS, kernel)
np.unique(oFour, return_counts=True)
```

```python
o = np.logical_or(np.logical_or(oOne, oTwo), np.logical_or(oThree, oFour))
points = []
for i in range(o.shape[0]):
    for j in range(o.shape[1]):
        if o[i, j]: 
            points.append((i, j))
points = np.array(points)
```

b) How did you find those pixels? What inputs did you use? Explain your process with details.

To find the corner pixels, morphological operations were applied. The goal was to look for fits using a cv2 function.

To find the corners in all directions, different kernels were used. Each one of them looking for a different kind of corner.

At the end, the results are tallied by applying a logical and then scattered over the image.

As we can see, it's able to spot really well corners with angles greater than 90 degrees.
