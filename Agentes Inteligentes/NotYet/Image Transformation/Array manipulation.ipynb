```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("MyPic.jpeg")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

## 1. Mirror image

```python
mirror = np.flip(img,axis=0)
plt.imshow(cv2.cvtColor(mirror, cv2.COLOR_BGR2RGB))
plt.show()
```

## 2. Show  horizontal half top image

```python
def horizontal_half(img):
    half = np.resize(img, [int(img.shape[0]/2), img.shape[1], img.shape[2]])
    newhalf = half.copy()
    newhalf.resize([int(img.shape[0]), img.shape[1], img.shape[2]])
    half = np.roll(newhalf, int(img.shape[0]/2), axis=0)
    return half

plt.imshow(cv2.cvtColor(horizontal_half(img), cv2.COLOR_BGR2RGB))
plt.show()
```

## 3. Show vertical half top image

```python
def vertical_half(img):
    half = np.roll(img, int(img.shape[1]/2), axis=1)
    for i in range(int(img.shape[1]/2)):
        for j in range(img.shape[0]):
            half[j,i] = [0,0,0]
    return half
plt.imshow(cv2.cvtColor(vertical_half(img), cv2.COLOR_BGR2RGB))
plt.show()
```

## 4. Translate h/2 in y & w/2 in x

```python
plt.imshow(cv2.cvtColor(horizontal_half(vertical_half(img)), cv2.COLOR_BGR2RGB))
plt.show()
```
