```python
import numpy as np
import pandas
from sklearn.datasets import load_digits
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
from sklearn.manifold import MDS
import matplotlib.pyplot as plt
```

# [Read the dataset](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html)

```python
x, y = load_digits(return_X_y = True)
print('x:', x.shape, '\ny:', y.shape)
classes = load_digits().target_names
print(classes)
```

# Principal Components Analysis

```python
modelPCA = PCA()
modelPCA.fit(x)
dataPCA = modelPCA.transform(x)
dataPCA.shape
```

## Plot the explained variance ratio of the components. 

```python
expVar = modelPCA.explained_variance_ratio_
print(np.round(expVar, 3))
expVarCum = np.cumsum(expVar)
plt.figure()
plt.plot(expVarCum)
plt.show()
```

## Print the digits map using the first two components
*Add the explained variance to the axis.*



```python
dataPCA = dataPCA[:, :2]
plt.figure(figsize=(10,10))
plt.scatter(dataPCA[:, 0], dataPCA[:, 1], s=1)
for i in range(50):
    plt.text(dataPCA[i, 0], dataPCA[i, 1], y[i])
plt.xlabel(np.round(expVar[0], 3)*100)
plt.ylabel(np.round(expVar[1], 3)*100)
plt.show()
```

# Linear Discriminant Analysis

```python
modelLDA = LinearDiscriminantAnalysis()
modelLDA.fit(x, y)
dataLDA = modelLDA.transform(x)
dataLDA.shape
```

## Print the digits map using the first two components


```python
plt.figure(figsize=(10,10))
plt.scatter(dataLDA[:, 0], dataLDA[:, 1], s=1)
for i in range(50):
    plt.text(dataLDA[i, 0], dataLDA[i, 1], y[i])
plt.show()
```

# Multidimensional Scaling

```python
modelMDS = MDS(n_components = 2) 
dataMDS = modelMDS.fit_transform(x)
dataMDS.shape
```

## Print the digits map using the first two components


```python
plt.figure(figsize=(10,10))
plt.scatter(dataMDS[:, 0], dataMDS[:, 1], s=1)
for i in range(50):
    plt.text(dataMDS[i, 0], dataMDS[i, 1], y[i])
plt.show()
```
