```python
import pandas
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
from sklearn.svm import SVR
```

```python
#df = pandas.read_csv('../Datasets/202201010000Lluv.csv', encoding = 'ISO-8859-1')
df = pandas.read_csv('https://raw.githubusercontent.com/LuisR-jpg/School/master/Machine%20Learning%20I/Datasets/202201010000Lluv.csv', encoding = 'ISO-8859-1')
df.head()
```

```python
plt.figure()
plt.subplot(1, 2, 1)
plt.boxplot(df.values[:, -1])
plt.subplot(1, 2, 2)
plt.boxplot(df.values[:, -1])
plt.ylim([0, 100])
plt.show()
```

```python
plt.figure()
plt.scatter(df['LON'].values, df['LAT'].values, s = 2, c = df.values[:, -1], vmin = 0, vmax = 40, cmap = 'jet')
plt.colorbar()
plt.show()
```

```python
Xtrain = df[['LON', 'LAT']].values
Ytrain = df.values[:, -1]
print('Xtrain', Xtrain.shape, '\nYtrain', len(Ytrain))
```

```python
model = SVR()
model.fit(Xtrain, Ytrain)
```

```python
minx,miny = np.min(Xtrain,axis=0)
maxx,maxy = np.max(Xtrain,axis=0)

N = 100
rowx = np.linspace(minx, maxx, N)
rowy = np.linspace(miny, maxy, N)
xx, yy = np.meshgrid(rowx, rowy)
zz = np.zeros((N, N), float)
color = np.zeros((N,N,4))

norm = matplotlib.colors.Normalize(vmin=0, vmax=40)
for x in range(N):
    for y in range(N):
        vec = np.array([[ xx[x, y], yy[x, y] ]])
        zz[x,y] = model.predict(vec)
        c = list(matplotlib.cm.jet(norm(zz[x, y])))
        color[N-1 - x, y, :] = c
color = (color * 255).astype(int)
```

```python
plt.figure()
plt.scatter(df['LON'].values,df['LAT'].values,s=2,c=df.values[:,-1],vmin=0,vmax=40,cmap='jet')
plt.imshow(color,extent=(minx,maxx,miny,maxy),alpha=0.5)
plt.xticks([])
plt.yticks([])
plt.xlim([minx, maxx])
plt.ylim([miny, maxy])
plt.show()
```

# Landmarks

```python
nL = 100
n = len(Xtrain)
idx = np.random.permutation(n)[: nL]
XL = Xtrain[idx, :]

plt.figure()
plt.scatter(Xtrain[:, 0], Xtrain[:, 1], s = 1)
plt.scatter(XL[:, 0], XL[:, 1], s = 5)
plt.show()
```

```python
XLtrain = np.zeros((n, nL))
for i in range(n):
    for j in range(nL):
        XLtrain[i, j] = np.linalg.norm((Xtrain[i, :] - XL[j, :]))
print('Xtrain', Xtrain.shape)
print('XLtrain', XLtrain.shape)
```

```python
model = SVR(kernel = 'linear')
model.fit(XLtrain, Ytrain)
```

```python
from shapely.geometry import Point
from geopandas import GeoSeries
from shapely.geometry import Polygon
import geopandas as gpd
mundi = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
mx = mundi[mundi['name'] == 'Mexico']
mx.plot()
mx['geometry']
```

```python
minx,miny = np.min(Xtrain,axis=0)
maxx,maxy = np.max(Xtrain,axis=0)

N = 100
rowx = np.linspace(minx, maxx, N)
rowy = np.linspace(miny, maxy, N)
xx, yy = np.meshgrid(rowx, rowy)
zz = np.zeros((N, N), float)
color = np.zeros((N,N,4))
norm = matplotlib.colors.Normalize(vmin=0, vmax=40)
for x in range(N):
    for y in range(N):
        vec = np.array([[ xx[x, y], yy[x, y] ]])
        vecL = np.zeros((1, nL))
        for j in range(nL):
            vecL[0, j] = np.linalg.norm((vec - XL[j, :]))
        zz[x,y] = model.predict(vecL)
        c = list(matplotlib.cm.jet(norm(zz[x, y])))
        pAux = Point(xx[x,y], yy[x, y])
        if mx['geometry'].contains(pAux).all(): color[N-1 - x, y, :] = c
        else: color[N-1 - x, y, :] = [0, 0, 0, 0]
color = (color * 255).astype(int)
```

```python
plt.figure()
plt.scatter(df['LON'].values,df['LAT'].values,s=2,c=df.values[:,-1],vmin=0,vmax=40,cmap='jet')
plt.imshow(color,extent=(minx,maxx,miny,maxy),alpha=0.5)
plt.xticks([])
plt.yticks([])
plt.xlim([minx, maxx])
plt.ylim([miny, maxy])
plt.show()
```
